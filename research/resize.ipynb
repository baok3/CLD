```python
import os
from PIL import Image
```

```python
def setup_directories():
    """Tạo các thư mục nếu chưa tồn tại và resize ảnh về kích thước 224x224"""
    base_dir = './data/nematode'
    dirs = ['train/image', 'test/image', 'val/image']
    
    # Tạo thư mục
    for d in dirs:
        dir_path = os.path.join(base_dir, d)
        os.makedirs(dir_path, exist_ok=True)
        
        # Kiểm tra và resize ảnh trong thư mục nếu tồn tại
        if os.path.exists(dir_path):
            for img_name in os.listdir(dir_path):
                if img_name.endswith(('.jpg', '.jpeg', '.png')):
                    img_path = os.path.join(dir_path, img_name)
                    try:
                        img = Image.open(img_path)
                        img_resized = img.resize((224, 224))
                        img_resized.save(img_path)
                        print(f"Đã resize thành công ảnh: {img_name}")
                    except Exception as e:
                        print(f"Lỗi khi xử lý ảnh {img_name}: {str(e)}")

setup_directories()
```
