```python
import os
from PIL import Image
```

```python
def resize_image():
    """Tạo các thư mục nếu chưa tồn tại và lưu kích thước gốc của ảnh"""
    base_dir = './data/rare_disease'
    dirs = ['Nematode', 'Pink Disease']
    
    # Tạo thư mục
    for d in dirs:
        dir_path = os.path.join(base_dir, d)
        os.makedirs(dir_path, exist_ok=True)
        
        # Kiểm tra và khôi phục kích thước gốc của ảnh trong thư mục nếu tồn tại
        if os.path.exists(dir_path):
            for img_name in os.listdir(dir_path):
                if img_name.endswith(('.jpg', '.jpeg', '.png')):
                    img_path = os.path.join(dir_path, img_name)
                    try:
                        img = Image.open(img_path)
                        original_size = [640, 360]
                        img_original = img.resize(original_size)
                        img_original.save(img_path)
                        print(f"Đã khôi phục kích thước gốc cho ảnh: {img_name}")
                    except Exception as e:
                        print(f"Lỗi khi xử lý ảnh {img_name}: {str(e)}")

resize_image()
```
